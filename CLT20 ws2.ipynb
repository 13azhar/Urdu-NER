processing (mini-)corpus of religious text

```
import gensim
import nltk

reader = nltk.corpus.PlaintextCorpusReader("/content/drive/My Drive/CLT20/Corpus",'.*\.txt')
text = reader.raw()
print(reader.fileids())
```


Text is split into sentences



```
import re

sentences = re.split(r'[۔؟]',text)

print("Count of Setnences: ",len(sentences))
print(sentences[6])
```

Sentences are split inot words.

```
from nltk.tokenize import WordPunctTokenizer
word_punct_tokenizer = WordPunctTokenizer()


s = []
for i in range(len(sentences)):
    s.append(word_punct_tokenizer.tokenize(sentences[i]))

print(s[6])
```

Trainaing Word2Vec model

```
model = gensim.models.Word2Vec(s, min_count=5, size = 20)
```

Printing related words

```
model.wv.most_similar('سورج')
```

```
model.wv.most_similar('فجر')
```
